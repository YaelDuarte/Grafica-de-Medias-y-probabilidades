```python
df_mx_microsoft = df_mx[df_mx["product_title"].str.contains("Microsoft", case = False, na = False)]


media_precio = df_mx_microsoft["product_price"].astype(str).str.replace(r"[^\d.]", "", regex=True).replace("", None).astype(float).mean()
media_rating = df_mx_microsoft["product_star_rating"].mean()
media_num_ratings = df_mx_microsoft["product_num_ratings"].mean()

print("üìä Medias para productos Microsoft en M√©xico:")
print(f"Precio promedio: {media_precio:.2f}")
print(f"Calificaci√≥n promedio: {media_rating:.2f}")
print(f"N√∫mero de rese√±as promedio: {media_num_ratings:.2f}")
```
